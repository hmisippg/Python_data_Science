```python
# reshaping data
import pandas as pd
```

```python
# pivoting Data => column -> baris
data = {
    'name' : ['naruto']*3 + ['sasuke']*3,
    'bar'  : ['A', 'B', 'C', 'A', 'B', 'C'],
    'bas'  : [1,2,3,4,5,6],
    'alpa' : ['d','e','f', 'g', 'h', 'i']
}

df = pd.DataFrame(data)
df
```

```python
# pivoting
pivoting_result = df.pivot_table(values='bas',index='name',columns='bar')
pivoting_result
```

```python
# Melting Data
df2 = pd.DataFrame({
    'Name'  : {0:'Rikudo', 1:'Sasuke', 2: 'Sakura'},
    'Class' : {0:'Sennin', 1:'Chunnin', 2:'Chunnin'},
    'Age'   : {0: 70, 1:25, 2:25}
})
df2
```

```python
# metling
pd.melt(df2, id_vars= ['Name'], value_vars= ['Class', 'Age'])
```
