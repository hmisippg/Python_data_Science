```python
import pandas as pd
```

```python
data = {
    'foo' : ['naruto']*3 + ['sasuke']*3,
    'bar' : ['A', 'B', 'C', 'A', 'B', 'C'],
    'baz' : [1, 2, 3, 4, 5, 6],
    'zoo' : ['x', 'y', 'z', 'q', 'w', 't']
}
df = pd.DataFrame(data)
df
```

```python
pivot_result = df.pivot_table(values='baz', index='foo', columns='bar')
pivot_result
```
